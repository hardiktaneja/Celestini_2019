Implemented DNN Regressor (inbuilt) . 

```python
import tensorflow as tf
```

```python
df = pd.read_csv("/Users/hardiktaneja/Downloads/zoo_data.csv")
```

```python
feature_columns=[]
```

```python
for key in df.keys():
    if(key == 'animal_name'):
        continue
    elif(key == 'type'):
        continue
    feature_columns.append(tf.contrib.layers.feature_column.real_valued_column(key))
```

```python
Classifier = tf.estimator.DNNClassifier(feature_columns = feature_columns,hidden_units=[10,10],n_classes=8)
```

```python
y = df["type"]
```

```python
x = df.drop(["animal_name","type"],axis=1)
```

```python
x_train,x_test,y_train,y_test = sklearn.model_selection.train_test_split(x,y,test_size=.2)
x_train = dict(x_train)
x_test = dict(x_test)
```

```python
def f1():
    return x_train,y_train
```

```python
def f2():
    return x_test,y_test
```

```python
Classifier.train(input_fn = f1,steps= 200)
```

95% accuracy achieved.

```python
eval_result = Classifier.evaluate(input_fn =f2,steps=1)
print(eval_result)
```
